```python
### 히트맵을 통해서 상관관계 파악
plt.figure(figsize = (12,10))
df_co = df.corr()
sns.heatmap(df_co,annot=True, annot_kws = {'size':12})
```

```python
# 연관성이 높은 특정 칼럼만 확인해보기
ccol = ['전체차량추정교통량','승용차추정교통량','혼잡시평균속도','차량평균속도']
sns.pairplot(df[ccol])
```

```python
corr2 = df[ccol].corr()
sns.heatmap(corr2,annot=True, annot_kws={'size':13})
```

```python
X = df.drop('정상시평균속도',axis=1)
y = df.정상시평균속도
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def eval_score(y_test,y_pred):
    mse=mean_squared_error(y_test,y_pred)
    rmse=np.sqrt(mse)
    r2=r2_score(y_test,y_pred)
    
    print('MSE : ',np.round(mse,3))
    print('RMSE : ',np.round(rmse,3))
    print('R2 : ',np.round(r2,3))
```

```python
eval_score(y_test,y_pred)
```

```python
### 회귀계수 확인

np.set_printoptions(suppress=True)

print('y절편 : {:.3f}'.format(lr.intercept_))
print('회귀 계수 : ',np.round(lr.coef_,3))
# ndarray의 실수 출력 옵션 복구
np.set_printoptions(suppress=False)
```

```python
#회귀  모델에서 구한 회귀 계수와 피처 이름을 묶어 시리즈 자료형으로 만들고,
coef=pd.Series(data=np.round(lr.coef_,3),index=X.columns)
#회귀 계수값을 기준으로 내림차순으로 정렬
coef_sort = coef.sort_values(ascending=False)
```

```python
#회귀  모델에서 구한 회귀 계수와 피처 이름을 묶어 
#시리즈 자료형으로 만듦
print('y = '+str(lr.intercept_) + ' +  ' )

for i,c in zip(coef.index,coef.values):
    print('(' + str(c) + ') * ' + str(i) + " + ")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(y_test,y_pred,s=80) # s : 점의 크기
plt.plot([30,70],[30,70],c='r',ls='--') # 실제값=예측값일때
plt.xlabel('Actual Velocity')
plt.ylabel('Predicted Velocity')
plt.grid()
plt.show()
```

```python
# 최소 제곱법으로 객체 생성, 학습, 예측, 평가를 진행한다.

import statsmodels.api as sm

# 훈련 데이터로 학습 수행
X_train=sm.add_constant(X_train)
reg=sm.OLS(y_train,X_train).fit()

# 검증 데이터로 예측 수행
X_test=sm.add_constant(X_test)
y_pred=reg.predict(X_test)

#평가 지표 값 확인
print(reg.summary())
```
